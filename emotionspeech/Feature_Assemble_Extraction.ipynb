```python
import pandas as pd
import librosa
import warnings
warnings.filterwarnings('ignore')

path="C:/Users/acer/Pictures/Speech_Processing/webinar/audio/"
```

### Feature Extraction Neutral

```python
dataset_neutral1 = pd.read_csv(path + "audio_neutral.csv") 
dataset_neutral1.head()
```

```python
audio_spec_neutral = []
for index, row in dataset_neutral1.iterrows():
    try:
        path_wav = path + 'potongan_neutral/' + row['nama_file'] 
        print(path_wav)
        audio_spec_neutral.append([librosa.load(path_wav),row['nama_file'],row['label']])
    except Exception as e:
        print("Error Baca ")
```

```python
data_extraction_neutral = []
def extract_features(signal):
    zcr = librosa.feature.zero_crossing_rate(signal[0][0])[0,0]
    sc  =librosa.feature.spectral_centroid(signal[0][0])[0,0]
    rmse = librosa.feature.rms(signal[0][0])[0,0]
    chroma = librosa.feature.chroma_stft(signal[0][0])[0,0]
    specto = librosa.feature.melspectrogram (signal[0][0])[0,0]
    return zcr, sc, rmse, chroma,specto
for x in audio_spec_neutral:
    try:
        data_extraction_neutral.append(extract_features(x))
        print("berhasil baca data")
    except:
        print("Error Baca File")
```

```python
dataset_neutral2 = pd.DataFrame(data=data_extraction_neutral, columns=['ZCR','SC','RMSE','Chroma', 'Spectogram'])
dataset_neutral2
```

```python
df_fix1 = dataset_neutral1.join(dataset_neutral2)
df_fix1 = df_fix1.drop(['nama_file'], axis=1).dropna()
df_fix1.head()
```
