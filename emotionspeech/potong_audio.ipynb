```python
from pydub import AudioSegment

path="C:/Users/acer/Pictures/Speech_Processing/webinar/audio/"
audio=AudioSegment.from_mp3(path+"audio-happy.mp3")
n=len(audio)
counter=1
interval=5*1000
overlap=0.0*1000
start=0
end=0
flag=0
for i in range(0,n,interval):
    if i == 0:
        start = 0
        end=interval
    else:
        start=end-overlap 
        end=start+interval
    chunk=audio[start:end]
    filename=path+'potongan_happy/pot'+str(counter)+'.wav'
    chunk.export(filename,format="wav")
    print(str(counter)+". Start = "+str(start/1000)+" \
        Detik end = "+str(end/1000) + " Detik")
    counter = counter + 1
```
