```python
import pandas as pd
import librosa
import warnings
warnings.filterwarnings('ignore')

path="C:/Users/acer/Pictures/jurnal/audio/"
```

### Feature Extraction Sound Campursari

```python
dataset_campursari1 = pd.read_csv(path + "audio_campursari.csv") 
dataset_campursari1.head()
```

```python
audio_spec_campursari = []
for index, row in dataset_campursari1.iterrows():
    try:
        path_wav = path + 'potongan_campursari/' + row['nama_file'] 
        print(path_wav)
        audio_spec_campursari.append([librosa.load(path_wav),row['nama_file'],row['label']])
    except Exception as e:
        print("Error Baca ")
```

```python
data_extraction_campursari = []
def extract_features(signal):
    zcr = librosa.feature.zero_crossing_rate(signal[0][0])[0,0]
    sc  =librosa.feature.spectral_centroid(signal[0][0])[0,0]
    rmse = librosa.feature.rms(signal[0][0])[0,0]
    chroma = librosa.feature.chroma_stft(signal[0][0])[0,0]
    specto = librosa.feature.melspectrogram (signal[0][0])[0,0]
    return zcr, sc, rmse, chroma,specto
for x in audio_spec_campursari:
    try:
        data_extraction_campursari.append(extract_features(x))
        print("berhasil baca data")
    except:
        print("Error Baca File")
```

```python
dataset_campursari2 = pd.DataFrame(data=data_extraction_campursari, columns=['ZCR','SC','RMSE','Chroma', 'Spectogram'])
dataset_campursari2
```

```python
df_fix1 = dataset_campursari1.join(dataset_campursari2)
df_fix1 = df_fix1.drop(['nama_file'], axis=1).dropna()
df_fix1.head()
```
